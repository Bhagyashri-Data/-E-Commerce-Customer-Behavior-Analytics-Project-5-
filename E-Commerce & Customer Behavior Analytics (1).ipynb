```python
#Task 18 Write an SQL query to retrieve the Name, City, and total count of orders for each customer (exactly like Task 17). However, filter the final output to display only those customers who have placed more than 1 order (i.e., Total_Orders > 1). Use the HAVING clause right after the GROUP BY clause to apply this aggregate filter.
```

```python
result2 = """
SELECT customer.Name,customer.City,COUNT(orders.order_date) AS 'Total_orders' FROM customer
LEFT JOIN orders ON customer.Cust_id = orders.Cust_id
GROUP BY customer.Cust_id
HAVING COUNT(orders.order_date) > 1
ORDER BY Total_Orders DESC;
"""
```

```python
result18 = pd.read_sql(result2,cust_conn)
result18
```
